## Building a Conversational Model

```python
#from the transformer library import the pipeline function
from transformers import pipeline
```

```python
#create a chatbot by passing  arguments in the pipeline function (task, model)
chatbot = pipeline(task ='conversational', model='facebook/blenderbot-400M-distill')
```

```python
#create an input message to use in the model
user_message = """
What are some fun activities I can do in the winter?
"""
```

```python
#from transformers import conversation class
from transformers import Conversation
```

```python
#pass the user message as argument for conversation class
conversation = Conversation(user_message)

print(conversation)
```

```python
#pass user message in the conversation object to the chatbot to get response from the model
conversation = chatbot(conversation)

print(conversation)
```

```python
#try it with another follow up message
user_message = "I like going on hiking"

conversation = Conversation(user_message)

conversation = chatbot(conversation)

print(conversation)
```

```python
#try it with another follow up message
#the output doen't show that the model remembers
#what we're talking about from above, as it has lost memory
print(chatbot(Conversation('What else do you recommend?')))
```

```python
#totally off-track here
print(chatbot(Conversation('her! who are you talking about?')))
```

```python
print(chatbot(Conversation('Oh! I see!')))
```

```python
#To ensure that the model stays on track use this code template instaed
conversation.add_message({'role': 'user',
    'content': """ What else do you recommend? """})
```

```python
#This output shows that it remembers the context of our discussion above.
print(conversation)
```
